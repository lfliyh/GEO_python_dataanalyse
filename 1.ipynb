```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

# 读取Excel文件
file_path = 'D:\\py\\530\\data\\Lab2_granite_compression.xlsx'  # 替换为你的Excel文件路径
sheet_name = 'Sheet2'  # 替换为你的工作表名称
data = pd.read_excel(file_path, sheet_name=sheet_name)

# 提取数据
strain = data["应力"]
axial_strain = data["横向应变"]
lateral_strain = data["轴向应变"]
num_of_ae_events = data["累计数"]

# 创建图形
fig, ax1 = plt.subplots()

# 绘制应力-应变曲线
ax1.plot(axial_strain,strain, label='axial_strain', color='orange')
ax1.plot( lateral_strain,strain, label='lateral_strain', color='blue')
ax1.set_xlabel('strain')
ax1.set_ylabel('stress')
ax1.spines['left'].set_position(('data', 0))  # 将 y 轴移动到 x=0
ax1.set_xlim(-0.2,0.5)
ax1.set_ylim(0, 250)
ax1.legend(loc='upper left')
#去掉多余的坐标轴
ax1.spines['top'].set_visible(False)

#创建第二个y轴
ax2 = ax1.twinx()
ax2.fill_between(lateral_strain,num_of_ae_events/100, color='green', alpha=0.5)
ax2.plot(lateral_strain,num_of_ae_events/100, label='Num of AE Events(00)', color='green')
ax2.set_ylabel('Num of AE Events(00)')
ax2.set_ylim(0, 250)
ax2.yaxis.set_major_locator(MultipleLocator(50))
#去掉多余的坐标轴
ax2.spines['left'].set_visible(False)
ax2.spines['top'].set_visible(False)


#合并标注
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc="upper right")


##设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置中文字体为SimHei
plt.rcParams['axes.unicode_minus'] = False  # 解决负号'-'显示为方块的问题


plt.title('Stress-Strain and AE Events 202210155李云豪')
#保存图形
plt.savefig('D:\\py\\530\\fesult\\Stress-Strain and AE Events 202210155李云豪.png')
#显示图形
plt.show()


```
